```julia
issymmetric(Eco4)
```

```julia
function recover(obs,lambda,n,rho,err,iter)
    recv = admmRecovery(obs,lambda,n,rho,err,iter)
    RealRecov=(recv[1]+ones(n)*ones(n)')/2
    recvRounded = round.(RealRecov)
    #checks symmetry
    if(issymmetric(recvRounded[1:n,1:n])==false)
        return("ERROR, non symetric matrix recovered.")
    end
    g = Graph(recvRounded)
    conected = connected_components(g)
    maximal_cliques(g)
```

```julia
##### recover the clusters...
recover=admmRecovery(EcoSenadores,0.001,156,3,1.0e-3,1000)
```

```julia
#real matrix
RealRecovery=(recover[1]+ones(156)*ones(156)')/2
```

```julia
recuperaRedondeada = round.(RealRecovery)
```

```julia
issymmetric(recuperaRedondeada[1:156,1:156])
```
