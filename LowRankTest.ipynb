<h1>Pruebas de recuperacion espectral<h1>

<h4>Para todas las pruebas de la recuperacion espectral, se suma un 1 en la diagonal.<h4>

```julia
verdadera = erdosgraph([1,1],0,[10,10],20)+diagm(ones(20))
```

```julia

grafo20obs3 = Array{Any}(3)
grafo20obs3[1]=erdosgraph([0.6,0.6],0.1,[10,10],20)+diagm(ones(20))
grafo20obs3[2]=erdosgraph([0.6,0.6],0.1,[10,10],20)+diagm(ones(20))
grafo20obs3[3]=erdosgraph([0.6,0.6],0.1,[10,10],20)+diagm(ones(20))
1
```

```julia
respuesta1 = learnEspectralGraph([1/3,1/3,1/3],0.2,grafo20obs3,20)
```

```julia
round.(respuesta1[1],2)
```

```julia
distance1(verdadera,respuesta1[1])
```

```julia
respuestalow=learnLowRankEspectral([1/3,1/3,1/3],0.2,grafo20obs3,20,0)
```

```julia
round.(respuestalow[1],2)
```

```julia
distance1(verdadera,respuestalow[1])
```

```julia
errores = Array{Any}(101)
constante = Array{Any}(101)
contador=1
for  t in 0:0.05:5
    constante[contador]= distance1(verdadera,respuesta1[1])
    recupera = learnLowRankEspectral([1/3,1/3,1/3],0.2,grafo20obs3,20,t)[1]
    errores[contador] = distance1(verdadera,recupera)
    contador = contador+1
end
    
    
```

```julia
#plot(0:0.05:5,P[1:30,1],label="delta=0.01")
plot(0:0.05:5,errores,label="RecuperacionBajoRango")
plot(0:0.05:5,constante,label="RecuperacionUsual")
grid("on")
legend(bbox_to_anchor=[1.05,1],loc=2,borderaxespad=0)
```
