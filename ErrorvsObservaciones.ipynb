<h2>Aprendizaje con norma 1<h2>

```julia
tic()
P = evaluarLinearGraphLearning6(0.7,0.3,0.01)
for s in [0.2,0.5,0.8,1.1,1.5,2]
    A = evaluarLinearGraphLearning6(0.7,0.3,s)
    P = [P A]
    println(s)
end
toc()
```
