```julia
papa = 0:0.01:8
```

```julia
alpha = 0.25
```

```julia
n = 50
```

```julia
func=function(delta)
    return(delta*(alpha-delta/n)^2/(1+2*delta/n))
end
```

```julia
resultado= map(func,papa
)
```

```julia
plot(papa,resultado)
grid("on")
legend(bbox_to_anchor=[1.05,1],loc=2,borderaxespad=0)
```
